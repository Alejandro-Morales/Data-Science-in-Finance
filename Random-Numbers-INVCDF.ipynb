```python
import scipy.optimize
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gaussian Random Numbers
f = lambda x, u: (1+scipy.special.erf((x-m)/np.sqrt(2*s**2)))/(2) - u
# viene de C^-1(U) = y >  C^-1(U) - y = 0 > aplicando funcion C > C(C^-1(U)) - C(y) = C(0). [C(0) = 0 obvi]
# = U - C(y) = 0 ======= C(y) - U = 0

n_rand = 5000
Gaus_random_numbers = np.zeros(n_rand)
m = 0
s = 1

for i in range(n_rand):
    u = np.random.uniform()
    #Nuestra busqueda de cero empezo en x = 0, cuidado cambiando m chance no funcionara
    y = scipy.optimize.fsolve(f,0,args = (u)) 
    Gaus_random_numbers[i] = y
```

```python
NB = 50
h,xr = np.histogram(Gaus_random_numbers,NB, density = True)
x = np.linspace(m-5*s,m+5*s,1000)
y = np.exp(-(x-m)**2/(2*s**2))/np.sqrt(2*np.pi*s**2)

plot0 = plt.figure(0,figsize=(7,4), dpi=100)
plt.title('Gauissian Random Numbers')
plt.plot(xr[:-1],h, "black")
plt.plot(x,y, "g")
```

```python
# Exponential Random Numbers
f = lambda x, u: (1 - np.exp(-lam*x)) - u

lam = 0.25
n_rand = 5000
Exp_random_numbers = np.zeros(n_rand)

for i in range(n_rand):
    u = np.random.uniform()
    y = scipy.optimize.fsolve(f,5,args = (u)) # Igual que arriba, checa que valor inicial metes para que busque el 0
    Exp_random_numbers[i] = y
```

```python
NB = 50
h,xr = np.histogram(Exp_random_numbers,NB, density = True)
x = np.linspace(0,8/lam,1000)
y = lam*np.exp(-lam*x)

plot0 = plt.figure(0,figsize=(7,4), dpi=100)
plt.title('Exponential Random Numbers')
plt.plot(xr[:-1],h,"black")
plt.plot(x,y,"g")
```
